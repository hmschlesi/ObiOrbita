# DAW Connection of the Orbita

Below is a general guide on how to set up your Orbita, built on a Raspberry Pi Pico, on different Digital Audio Workstations (DAWs) such as Ableton, FL Studio, Logic Pro, Cubase, Reaper and GarageBand. Please note that the setup process may vary slightly depending on the specific DAW version and operating system (OS) you are using. 

Always refer to the official documentation of your DAW for detailed and up-to-date instructions. By consulting the official documentation and resources specific to your DAW version, you can follow the most up-to-date and accurate instructions for setting up your Orbita with the DAW of your choice. Below are links to the up to date (in August 2023) manuals of those DAWs:

FL Studio: <https://www.image-line.com/fl-studio-learning/fl-studio-online-manual/>

**Note** : Your Orbita MIDI device should be recognized automatically by your computer after you connect it via the provided USB cable.

## Setting up your Orbita with Ableton:

1. Connect your Orbita MIDI sequencer to your computer via USB.
1. Open Ableton and go to "Options" (Windows) or "Live" (Mac) in the top menu, then choose "Preferences."
1. In the "Preferences" window, click on the "MIDI" tab.
1. Under the "MIDI Ports" section, locate your Raspberry Pi MIDI device. It should be listed as an available MIDI input device.
1. Enable the input for your Raspberry Pi MIDI device by toggling the “Track” switch next to its name.
1. Close the "Preferences" window.
1. In Ableton, create a new MIDI track by right clicking on the workspace and choosing “Insert MIDI Track”.
1. On the newly created MIDI track, set the "MIDI From" input to your Orbita or leave it at “All Ins”.
1. Drag and drop a MIDI Instrument of your choice on the MIDI track.
1. Arm the MIDI track for recording to receive MIDI signals.
1. Your Orbita MIDI sequencer is now set up with Ableton. You should be able to record MIDI data from the sequencer onto Ableton's MIDI track.

## Setting up your Orbita with FL Studio:

1. Connect your Orbita MIDI sequencer to your computer via USB.
1. Open FL Studio and go to "Options" in the top menu, then choose "MIDI Settings."
1. In the "MIDI Settings" window, locate your Orbita MIDI device in the "Input" section. It should be listed as an available MIDI input device.
1. Enable the input for your Orbita by checking the box next to its name.
1. Close the "MIDI Settings" window.
1. In FL Studio, create a new MIDI track in the Step Sequencer view.
1. On the newly created MIDI track, set the "MIDI In" port to your Raspberry Pi MIDI device.
1. Arm the MIDI track for recording or enable it to receive MIDI signals.
1. Your Orbita is now set up with FL Studio. You should be able to record MIDI data from the sequencer onto FL Studio's MIDI track.

## Setting up your Orbita with Logic Pro:

1. Connect your Orbita MIDI sequencer to your computer via USB.
1. Open Logic Pro and go to "Logic Pro" in the top menu, then choose "Control Surfaces" and "MIDI Controllers."
1. In the "MIDI Controllers" window, click on the "+" icon to add a new control surface.
1. From the "MIDI Input" dropdown menu, select your Orbita MIDI device.
1. Close the "MIDI Controllers" window.
1. In Logic Pro, create a new Software Instrument track.
1. On the newly created track, set the "MIDI Input" to your Orbita MIDI device.
1. Arm the Software Instrument track for recording or enable it to receive MIDI signals.
1. Your Orbita is now set up with Logic Pro. You should be able to record MIDI data from the sequencer onto Logic Pro's Software Instrument track.

## Setting up your Orbita with Cubase:

1. Connect your Orbita MIDI sequencer to your computer via USB.
1. Open Cubase and go to "Studio" in the top menu, then choose "Studio Setup."
1. In the "Studio Setup" window, click on "MIDI Port Setup" on the left-hand side.
1. Under "MIDI Input," locate your Orbita MIDI device. It should be listed as an available MIDI input device.
1. Enable the input for your Orbita MIDI device by checking the box next to its name.
1. Close the "Studio Setup" window.
1. In Cubase, create a new MIDI track by clicking on "Project" in the top menu, then choose "Add Track" and "MIDI."
1. On the newly created MIDI track, set the "MIDI Input" to your Orbita MIDI device.
1. Arm the MIDI track for recording or enable it to receive MIDI signals.
1. Your Orbita is now set up with Cubase. You should be able to record MIDI data from the sequencer onto Cubase's MIDI track.

## Setting up your Orbita with Reaper:

1. Connect your Orbita MIDI sequencer to your computer via USB.
1. Open Reaper and go to "Options" in the top menu, then choose "Preferences."
1. In the "Preferences" window, click on "MIDI Devices" on the left-hand side.
1. Under "Input," locate your Orbita MIDI device. It should be listed as an available MIDI input device.
1. Enable the input for your Orbita MIDI device by checking the box next to its name.
1. Close the "Preferences" window.
1. In Reaper, create a new track by right-clicking on the track area and choosing "Insert New Track."
1. On the newly created track, set the "MIDI Input" to your Orbita MIDI device.
1. Arm the track for recording or enable it to receive MIDI signals.
1. Your Orbita is now set up with Reaper. You should be able to record MIDI data from the sequencer onto Reaper's MIDI track.

## Setting up your Orbita with GarageBand:

1. Connect your Orbita MIDI sequencer to your computer via USB.
1. Open GarageBand and create a new project.
1. Click on the gear icon in the top-right corner of the GarageBand window to open "Audio/MIDI Preferences."
1. Under the "MIDI" tab, locate your Orbita MIDI device. It should be listed as an available MIDI input device.
1. Enable the input for your Raspberry Pi MIDI device by checking the box next to its name.
1. Close the "Audio/MIDI Preferences" window.
1. In GarageBand, create a new Software Instrument track.
1. On the newly created track, set the "MIDI Input" to your Orbita MIDI device.
1. Arm the Software Instrument track for recording or enable it to receive MIDI signals.
1. Your Orbita is now set up with GarageBand. You should be able to record MIDI data from the sequencer onto GarageBand's Software Instrument track.

Please remember that your Orbita MIDI sequencer can only send MIDI signals, not receive them. Therefore, you won't be able to send MIDI data from your DAW to your sequencer. So you don’t have to activate the “Remote” or “Sync” switch.

With these instructions, you should be able to set up your Orbita MIDI sequencer with Ableton, FL Studio, Logic Pro, Cubase, Reaper and GarageBand successfully. Remember to consult the specific documentation of your DAW and operating system if you encounter any issues or need further assistance.
